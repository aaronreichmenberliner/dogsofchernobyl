# Figure 3B

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import matplotlib.ticker as ticker
```

```python
## Plot All Data
fig, ax = plt.subplots()
df = pd.read_excel('data1.xlsx', sheet_name='all')
df1 = pd.DataFrame(df).dropna()
# df1 = np.abs(pd.DataFrame(df).dropna())
df1 = df1[df1['Body Burden [Bq/kg]'] > 100]
ax.errorbar(df1['Mass [kg]'],df1['Body Burden [Bq/kg]'],yerr=df1['Body Burden Error [Bq/kg]'],marker='s',ls='none')
ax.set_yscale('log')
ax.set_xlabel('Dog Mass [kg]')
ax.set_ylabel(r'Estimated $^{137}$Cs Body Burden [Bq/kg]')
ax.set_ylim([10,15000])
```

```python
g = sns.jointplot(x='Mass [kg]',y='Body Burden [Bq/kg]', data=df1, palette='flare',legend=False, hue='Year', ax=ax, marker='o')
g.plot_joint(sns.scatterplot, data=df1[df1['cont']>1], size=df1[df1['cont']>1]['cont'], sizes=(20, 200), size_norm=(1, 2000), marker='s', legend=False)
# g.plot_joint(sns.plot, x=[0,45], y=[100,100],)
g.ax_joint.set_xlabel('Mass [kg]', fontsize=15)
g.ax_joint.set_ylabel('Body Burden [Bq/kg]', fontsize=17)
g.ax_joint.set_xlim([0,40])
g.ax_joint.set_yticklabels(g.ax_joint.get_yticks(), size = 15)
g.ax_joint.set_xticklabels(g.ax_joint.get_xticks(), size = 15)
g.ax_joint.set_yscale('log')
g.ax_joint.set_xscale('linear')
g.plot_joint(sns.kdeplot, zorder=0,legend=False, levels=5)
g.savefig('test.png', dpi=300, bbox_inches='tight')
```
